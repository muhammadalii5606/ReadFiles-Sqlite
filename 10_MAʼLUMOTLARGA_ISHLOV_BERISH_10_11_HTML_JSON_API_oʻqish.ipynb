#1-topshiriq
1. Pandas kutubxonasidan foydalanib, quyida berilgan link ichidan barcha jadvallarni o‘qing va roʻyxat shaklida saqlang.

Link 403 Forbidden (Manzil to'g'ri, lekin kirish taqiqlangan.) xatoligini bermasligi uchun, sahifani avval requests bilan yuklab, keyin pd.read_html() ga HTML matnini bering.

 * https://en.wikipedia.org/wiki/Fossil_fuel_exporters
2. Har bir jadvalning birinchi 1 qatorini ekranga chiqaring.


    

```python
import pandas as pd
import requests
from io import StringIO
url = "https://en.wikipedia.org/wiki/Fossil_fuel_exporters"
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
html_data = StringIO(response.text)
tables = pd.read_html(html_data)

print(f"Muvaffaqiyatli! Sahifada jami {len(tables)} ta jadval topildi.\n")

for i, df in enumerate(tables):
    print(f"--- {i}-jadval ---")
    print(df.head(1))
    print("\n")
```

#2-topshiriq
Yuqorida yaratilgan jadvallar roʻyxatidan eng uzunini `largest_data.csv` fayliga saqlang.

```python
largest_df = max(tables, key=len)
largest_df.to_csv('largest_data.csv', index=False)
print(f"Eng uzun jadval topildi!")
print(f"Uning uzunligi: {len(largest_df)} ta qator.")
print(f"Fayl 'largest_data.csv' nomi bilan saqlandi.")

print("\nJadvalning ko'rinishi:")
print(largest_df.head())
```

#4-topshiriq
1. Quyida berilgan linkga `get` soʻrovini yuborish orqali `json` maʼlumotlarini `posts.json` fayliga saqlang.
  * https://jsonplaceholder.typicode.com/posts/
2. Ushbu maʼlumotlarni dataframe koʻrinishida saqlang va tasodifiy 5 ta qatorni ekranga chiqaring.

```python
import json
url = "https://jsonplaceholder.typicode.com/posts/"
response = requests.get(url)
data = response.json()

with open('posts.json', 'w') as f:
    json.dump(data, f, indent=4)

df = pd.DataFrame(data)

print("Tasodifiy 5 ta qator (sample):")
print(df.sample(5))
```
