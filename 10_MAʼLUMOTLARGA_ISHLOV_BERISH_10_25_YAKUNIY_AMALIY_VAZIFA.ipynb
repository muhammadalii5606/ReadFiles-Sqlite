#**5) APIʼlar bilan ishlash**
1. Quyida berilgan API 17 mingdan ortiq kriptovalyutalarning dollarga nisbatan narxi haqidagi maʼlumotlarni saqlaydi. Siz esa endi:
  * API orqali barcha maʼlumotlarni yigʻib oling;
  * ularni dataframe koʻrinishida saqlang;
  * maʼlumotlar boshidan 5 ta qatorni ekranga chiqaring;
  * maʼlumotlar ichidan `bitcoin` ni izlab toping!

````python
import requests

# 1. API manzilini belgilaymiz
coins_url = "https://api.coingecko.com/api/v3/coins/list"

# 2. API orqali ma'lumotlarni yig'ib olamiz
response = requests.get(coins_url)
data = response.json()

# 3. Ma'lumotlarni DataFrame ko'rinishida saqlaymiz
df_coins = pd.DataFrame(data)

# 4. Ma'lumotlar boshidan 5 ta qatorni ekranga chiqaramiz
print("Dastlabki 5 ta qator:")
print(df_coins.head(5))
print("\n" + "="*30 + "\n")

# 5. Ma'lumotlar ichidan 'bitcoin' ni izlab topamiz
# Bunda biz 'id' yoki 'name' ustunidan foydalanishimiz mumkin
bitcoin_info = df_coins[df_coins['id'] == 'bitcoin']

print("Bitcoin haqidagi ma'lumot:")
print(bitcoin_info)
````

#**6) APIʼlar bilan ishlash**
1. Yuqorida berilgan kriptovalyuta maʼlumotlaridan `id` ustunini roʻyxat koʻrinishida saqlang va bu roʻyxatga `coin_ids` deb nom bering.
2. https://api.coingecko.com/api/v3/simple/price?ids={coin_ids}&vs_currencies=usd havolasi orqali kriptovalyutalarning `usd` ga nisbatini oling (birinchi 100 ta maʼlumotni olish yetarli).
3. 2-shartda berilgan link orqali olingan maʼlumotlarni dataframe koʻrinishida saqlang. Bunda uning indeksi va `usd` ustuni boʻlsin. `index` -> kriptovalyutalar nomi, `usd` -> narxi

#**💡Ishora:** linkdagi `{coin_ids}` oʻzgaruvchi sifatida yozilgan. Siz ham 1-shartda saqlagan `coin_ids` oʻzgaruvchingizni shu qismda yozib keting!


````python
coin_ids_list = df_coins['id'].head(100).tolist()

# API so'rovi uchun ro'yxatni "bitcoin,ethereum,tether..." ko'rinishidagi matnga aylantiramiz
coin_ids_query = ",".join(coin_ids_list)

# 2. Narxlarni olish uchun URL shakllantiramiz
price_url = f"https://api.coingecko.com/api/v3/simple/price?ids={coin_ids_query}&vs_currencies=usd"

# 3. So'rov yuboramiz va natijani olamiz
response = requests.get(price_url)
price_data = response.json()

# 4. Ma'lumotlarni DataFrame ko'rinishida saqlaymiz
# price_data formati: {'bitcoin': {'usd': 65000}, 'ethereum': {'usd': 3500}}
# Shuning uchun uni 'index' (qatorlar bo'yicha) ko'rinishida o'qiymiz
df_prices = pd.DataFrame.from_dict(price_data, orient='index')

# Natijani ko'ramiz
print("Kriptovalyutalar va ularning USD narxi:")
print(df_prices.head(10))
````

#**7) Veb sahifadan oʻqish**
1. Pandas yordamida veb sahifadagi jadvalni yuklab oling.
2. Ma’lumotlarni dataframe ko‘rinishida chiqaring.
3. Agar sahifada bir nechta jadval bo‘lsa, eng muhimini tanlang.
4. Ma’lumotni tozalang (bo‘sh qatorlarni olib tashlang, kerakli ustunlarni tanlang).
5. Natijani CSV faylga saqlang.

***Eslatma: manba sifatida oʻzingiz istagan veb sahifani tanlang. Bu vazifada mustaqil izlanishni oʻrganasiz!***

````python
from io import StringIO

# 1. Manba URL (Dunyoning eng yirik kompaniyalari ro'yxati)
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"

# 2. 403 xatoligini oldini olish uchun so'rov yuboramiz
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)

# 3. Sahifadagi jadvallarni o'qiymiz
html_data = StringIO(response.text)
tables = pd.read_html(html_data)

# Wikipedia sahifasida odatda 1- yoki 2-jadval asosiy bo'ladi
# Bizga "List of largest companies" jadvali kerak (odatda index 0 yoki 1)
df = tables[0]

# 4. Ma'lumotlarni tozalash
# a) Bo'sh qatorlarni o'chirib tashlaymiz
df.dropna(inplace=True)

# b) Kerakli ustunlarni tanlaymiz (Kompaniya nomi, Sanoat, Daromad, Davlat)
# Eslatma: Wikipedia jadvallarida ustun nomlari o'zgarishi mumkin, shuning uchun tekshirib olamiz
print("Asl ustunlar:", df.columns.tolist())

# Faqat dastlabki 5 ta muhim ustunni olamiz
df_cleaned = df.iloc[:, :5]

# 5. Natijani CSV faylga saqlash
df_cleaned.to_csv('top_companies.csv', index=False)

print("\nTozalangan ma'lumotlar (dastlabki 5 qator):")
print(df_cleaned.head())
print("\nMa'lumotlar 'top_companies.csv' fayliga muvaffaqiyatli saqlandi.")
````

#**9) Maʼlumotlar omboriga soʻrov yuborish**
1. Jadvallar roʻyxatini koʻring. Ularning ichida `albums` jadvali boʻlsa, shu jadvaldagi barcha maʼlumotlarni (qatorlarni) ekranga chiqaring.
2. Qatorlar soni nechta ekanini aniqlab, uni ham ekranga chiqaring.
3.`albums` jadvalining ustunlari haqidagi maʼlumotlarni ekranga chiqaring.


````python
# 1. Ma'lumotlar omboriga ulanish
db_connection = sqlite3.connect('chinook.db')

# 2. Jadvallar ro'yxatini ko'rish
tables_query = "SELECT name FROM sqlite_master WHERE type='table';"
all_tables = pd.read_sql(tables_query, db_connection)
print("Bazadagi jadvallar ro'yxati:")
print(all_tables)
print("-" * 30)

# 3. 'albums' jadvali borligini tekshirish va ma'lumotlarni o'qish
if 'albums' in all_tables['name'].values:
    # Barcha ma'lumotlarni o'qiymiz
    albums_df = pd.read_sql("SELECT * FROM albums", db_connection)

    print("'albums' jadvalidagi barcha ma'lumotlar:")
    print(albums_df)
    print("-" * 30)

    # 4. Qatorlar sonini aniqlash
    row_count = len(albums_df)
    print(f"Jadvaldagi jami qatorlar soni: {row_count}")
    print("-" * 30)

    # 5. Ustunlar haqida ma'lumot (turlari, bo'sh emasligi va h.k.)
    print("'albums' jadvalining ustunlari haqida ma'lumot:")
    print(albums_df.info())
else:
    print("Xatolik: 'albums' jadvali topilmadi!")

# Ulanishni yopish
db_connection.close()
````

#**15) CREATE and INSERT**
* `chinook.db` maʼlumotlar omboriga `users` jadvali yarating. Jadval maydonlari (ustunlar):
  * `id`
  * `name`
  * `birth_date`
* Yaratilgan `users` jadvaliga quyidagi maʼlumotlarni qoʻshing:

    * ```
    [
    ('Alice', 25), ('Bob', 30), ('Charlie', 35), ('David', 28), ('Eve', 22),
    ('Frank', 40), ('Grace', 27), ('Hannah', 32), ('Ivy', 29), ('Jack', 31),
    ('Kevin', 33), ('Liam', 26), ('Mia', 24), ('Noah', 36), ('Olivia', 23),
    ('Peter', 39), ('Quinn', 37), ('Rachel', 34), ('Samuel', 41), ('Tina', 21)
]
 ```

#💡Ishora: CREATE TABLE, INSERT


````python
db_connection = sqlite3.connect('chinook.db')
cursor = db_connection.cursor()

create_table_query = """
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER
);
"""
cursor.execute(create_table_query)

users_data = [
    ('Alice', 25), ('Bob', 30), ('Charlie', 35), ('David', 28), ('Eve', 22),
    ('Frank', 40), ('Grace', 27), ('Hannah', 32), ('Ivy', 29), ('Jack', 31),
    ('Kevin', 33), ('Liam', 26), ('Mia', 24), ('Noah', 36), ('Olivia', 23),
    ('Peter', 39), ('Quinn', 37), ('Rachel', 34), ('Samuel', 41), ('Tina', 21)
]

cursor.executemany("INSERT INTO users (name, age) VALUES (?, ?);", users_data)

db_connection.commit()
print(f"Jadval yaratildi va {len(users_data)} ta foydalanuvchi qo'shildi.")

import pandas as pd
print("\nYangi yaratilgan 'users' jadvali:")
print(pd.read_sql("SELECT * FROM users", db_connection))

db_connection.close()
````

#**16) UPDATE and DELETE**

````python
db_connection = sqlite3.connect('chinook.db')
cursor = db_connection.cursor()

update_query = "UPDATE users SET age = 28 WHERE name = 'Alice'"
cursor.execute(update_query)
print("Alice'ning yoshi yangilandi.")

delete_query = "DELETE FROM users WHERE name = 'Alice'"
cursor.execute(delete_query)
print("Alice ismli foydalanuvchi bazadan o'chirildi.")

db_connection.commit()

df_check = pd.read_sql("SELECT * FROM users LIMIT 5", db_connection)
print("\nYangilangan 'users' jadvali (dastlabki 5 qator):")
print(df_check)

db_connection.close()
````

#**18)  AND and OR operators**

* `invoices` jadvalidan `BillingCountry` Germaniya yoki Fransiya boʻlgan va `Total` summasi 5 dan katta bo‘lgan kerakli qatorlarni tanlang.

#💡Ishora: AND, OR

````python
db_connection = sqlite3.connect('chinook.db')

query = """
SELECT * FROM invoices
WHERE (BillingCountry = 'Germany' OR BillingCountry = 'France')
AND Total > 5
"""

df_filtered = pd.read_sql(query, db_connection)

print("Germaniya yoki Fransiyadagi 5 dan katta hisob-fakturalar:")
print(df_filtered[['InvoiceId', 'BillingCountry', 'Total']].head(10))

db_connection.close()
````
